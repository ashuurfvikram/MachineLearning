```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#We will find Gradient Descent for the function=f(x)=x^3 -3x^2+7
```

```python
#creating funtion and plotting it

function=lambda x:(x**3)-(3*(x**2))+7

#get 1000 evenly spaced numbers between -1 and 3(arbitrary chosen to ensure steep curve)
x=np.linspace(-1,3,1000)

#plot the curve
plt.plot(x,function(x))
plt.show()
```

```python
#we could see that minimum is around 2 lets find it out using Gradient Descent

def deriv(x):
    """
    It should return derivative of f(x)
    """
    
    x_deriv=3*(x**2)-(6*(x))
    return x_deriv


```

```python
def step(x_new,x_prev,precision,l_r):
    
    #create an empty list where updated value of x and y will be updated after every iteration
    x_list,y_list=[x_new],[function(x_new)]
    
    #loop with the condition when to stop
    while(abs(x_prev-x_new)>precision):
        
        #change the value of x 
        x_prev=x_new
        
        #get the derivation of old value of x
        d_x=-deriv(x_prev)
        
        #get your new value using the update function
        x_new=x_prev + (l_r*d_x)
        
        #append new values of x to a list of all new values of x's for later visualization
        x_list.append(x_new)
        
        #append new values of y to a list of all new values of function(x_new) for later visualization
        y_list.append(function(x_new))
        
    print("local minimum at: " +str(x_new))
    print("number of steps: "+ str(len(x_list)))
    
    plt.subplot(1,2,2)
    plt.scatter(x_list,y_list,c="g")
    plt.plot(x_list,y_list,c="g")
    plt.plot(x,function(x),c="r")
    plt.title=("Gradient Descent")
    plt.show()

    plt.subplot(1,2,1)
    plt.scatter(x_list,y_list,c='g')
    plt.plot(x_list,y_list,c='g')
    plt.plot(x,function(x),c='r')
    plt.xlim([1.0,2.2])
    plt.title("Zoomed in Gradient descent to key area")
    plt.show()

#Implementing gradient descent function

step(0.5,0,0.001,0.05)
    
    
```
